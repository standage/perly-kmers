```python
import random
import subprocess
import sys
```

```python
species = {
    'Bvul': 'Basiliscus_vulgaris',
    'Docc': 'Draconis_occidentalis',
    'Dori': 'Draconis_orientalis',
    'Emon': 'Equus_monoceros',
    'Epeg': 'Equus_pegasus',
    'Hdiu': 'Homo_diutinus',
    'Hhae': 'Homo_haemophilus',
}
```

```python
random.seed(1985)
numseqs = [random.randint(2, 5) for _ in range(len(species))]
seqlens = [[random.randint(150, 450) for i in range(numseqs[j])] for j in range(len(numseqs))]
seeds = [[random.randint(0, sys.maxsize) for i in range(numseqs[j])] for j in range(len(numseqs))]
print(numseqs)
print(seqlens)
print(seeds)
```

```python
for n, label in enumerate(species):
    binom = species[label]
    filename = binom + '.fasta'
    with open(filename, 'w') as outstream:
        for i in range(numseqs[n]):
            seqlen = seqlens[n][i]
            seed = seeds[n][i]
            name = '{}|seq{}'.format(label, i)
            cmd = ['./make-random-genome.py', '--seed', str(seed), '--name', name, '--length', str(seqlen)]
            subprocess.check_call(cmd, stdout=outstream)
```
